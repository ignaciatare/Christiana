```python
with open("vagina.txt") as reader:
    text_lines = reader.readlines()

text = " ".join(text_lines)
```

```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = stopwords.words('spanish') + [';', ':', ',', '"', '»', '“', '”', '«', ')', '$', '.', '%', '?', '(', '...', '.']

counter = Counter()

words = [word.lower() for word in word_tokenize(text) if word.lower() not in stop_words]
counter = Counter(words)
```

```python
import pandas as pd

df = pd.DataFrame(counter.items(), columns=["palabra", "cantidad"])
df
```

```python
df.sort_values("cantidad", ascending = False)
```

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer('spanish')

#df["tallo"] = df.palabra.apply(stemmer.stem) #nueva fila
#df.sort_values("cantidad", ascending = False) #ranking según cantidad, no tallo

words = [stemmer.stem(word.lower()) for word in word_tokenize(text) if word.lower() not in stop_words]

counter_stem = Counter(words)
counter_stem.most_common(20), counter.most_common(20)
```
